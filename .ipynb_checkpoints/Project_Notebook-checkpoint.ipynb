# Analyzing What Factors Affect Quality of Life

## Background
### What is the quality of life?
According to Oxford Languages, the quality of life is **"the standard of health, comfort, and happiness experienced by an individual or group."** According to the World Health Organization, the quality of life is **“an individual’s perception of their position in life” according to their cultural backgrounds, values, and lifestyles (Wikipedia).**

The quality of life is known to be both objective and subjective in this way: certain factors or "standards" in our environment are constant, such as how much pollution there is in the air, or how much education one recieves. However, the quality of life may also be defined as "perception," or what we see as a healthy, comfortable, and happy life based on what we think is important and prioritize. For example, some may think one's quality of life increases with better healthcare, while others may think their environment is more important.

The engaged theory may link both of these definitions together, as it establishes **"four domains in assessing quality of life: ecology, economics, politics and culture"** (Wikipedia). I will use this theory as a basis for categorizing the data and analyzing the importance of each factor in the quality of life.  

![Visualizing the quality of life](../Img/qol-identifier-eng.png)

## Goals and Signifance
By the end of this notebook, I hope to answer the following questions: 
Does the quality of life have **more correlation with one of the domains than another**, or are all domains **equally important**?
Are there **specific areas** within each domain that significantly impact a nation’s quality of life? 
And lastly, do we collectively emphasize one of these domains more than the other through **online sources** that talk about the quality of life?

By studying the quality of life, leaders can focus on areas that will positively impact the quality of life for citizens. We can learn more about **society's perspective** on what makes us happy, and create policies that could then help us lead happier lives.

### Quality Of Life Index Formulas

One of the columns in this data is labeled "Quality of Life Index." According to the creator of
this dataset, the Quality of Life Index came from Numbeo, based of off 2017 data.

In 2017, Numbeo used two **formulas to calculate the quality of life index**: 

From November 2015 to June 2017 : "index.main = Math.max(0, 100 + purchasingPowerInclRentIndex / 2.5 - (housePriceToIncomeRatio * 1.0) - costOfLivingIndex / 5 + safetyIndex / 2.0 + healthIndex / 2.5 - trafficTimeIndex / 2.0 - pollutionIndex * 2.0 / 3.0 + climateIndex / 2.0);"

From June 2017 to December 2017 : "index.main = Math.max(0, 100 + purchasingPowerInclRentIndex / 2.5 - (housePriceToIncomeRatio * 1.0) - costOfLivingIndex / 5 + safetyIndex / 2.0 + healthIndex / 2.5 - trafficTimeIndex / 2.0 - pollutionIndex * 2.0 / 3.0 + climateIndex / 3.0);" 

From these formulas, the columns that hold data such as safetyIndex and healthIndex will correlate. I would like to figure out what other factors of the data would correlate with the calculated quality of life, even if not explicitly used in the formula.

## Textual Data : Wikipedia Page

The numerical data is useful, but was more skewed towards economic data, such that it missed out on the domain of politics. I also wanted to look at textual data through online sources; **if you looked up what the quality of life was, would you think it was based on politics, economics, ecology, culture, or a little bit of everything?** In order to get a different perspective, I decided to use a Wikipedia page that defines the quality of life (https://en.wikipedia.org/wiki/Quality_of_life), and looked at all the other Wikipedia pages hyperlinked to the original page. Wikipedia pages often have key terms that users can click on to navigate to other resources, and often these terms are equally as important to understand. By looking at the other related Wikipedia pages, I had more textual data to work with and could find larger themes of the four domains in relation to the quality of life.
![Wikipedia Quality of Life Page](../Img/wiki_page.png)

### Getting the Links

```python
import requests
from bs4 import BeautifulSoup 
import time
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Quality_of_life')
wiki_html = page.text
s = BeautifulSoup(wiki_html)
text = s.find(id = "mw-content-text")
```

```python
#here I am appending all of the other Wikipedia pages linked to the original page to a list
links_list = []
links_list.append("https://en.wikipedia.org/wiki/Quality_of_life")
for line in text.find_all('a'):
    try:
       #print(line['href'])
       if(line['href'].find("wiki") != -1 & line['href'].find("https") == -1 & line['href'].find(":") == -1):
            if(line['href'].find(")") != -1):
                line['href'] = "https://en.wikipedia.org" + (line['href'])[0:-1] + "%29"
            else:
                line['href'] = "https://en.wikipedia.org" + line['href']
            if(links_list.count(line['href']) == 0): 
                links_list.append(line['href'])
    except KeyError:
        continue
links_list
```

```python
len(links_list)
```

```python
len(links_list) == len(set(links_list))
```

### Creating a Data Frame

Now that the Wikipedia links are ready, I will store them in a dataframe along with their content.

```python
#This step takes five minutes
df = {}
for link in links_list:
    page = requests.get(link)
    wiki_html = page.text
    s = BeautifulSoup(wiki_html)
    text = s.find(id = "mw-content-text")
    all_text = ""
    for i in text.find_all("p"):
        all_text = all_text + i.text
        #then add all_text to another column
    df[link] = all_text
    time.sleep(1)

wiki_data = (pd.DataFrame.from_dict(df, 
                                       orient = "index")
                .reset_index().rename(index = str, 
                                      columns = {"index": "Link", 0: "Text"}))
```

```python
wiki_data
```

```python
#some pages redirect to other pages with a different link, so this step makes sure that all values are unique
wiki_data.drop_duplicates(subset ="Text",
                     keep = 'first', inplace = True)
```

```python
wiki_data
```

### Preprocessing

```python
import gensim 
import numpy as np 
import os
import re
import gensim
from sklearn.decomposition import PCA
import warnings
from nltk.corpus import stopwords
from collections import Counter
warnings.filterwarnings("ignore", category = DeprecationWarning)
```

```python
#preprocessing
wiki_data["Processed Text"] = wiki_data["Text"].apply(gensim.utils.simple_preprocess)
wiki_data["Processed Text"]
```

```python
#From Notebook 5-2: Topic Modeling
processed = wiki_data["Text"].apply(gensim.utils.simple_preprocess)
stop = stopwords.words('english')
no_stop = processed.apply(lambda x: [w for w in x if w not in stopwords.words('english')])
wiki_data["Processed Text"] = [' '.join(t) for t in no_stop]
wiki_data
```

```python
wiki_data.to_csv('../Data/wiki_data.csv', index=False)
```

### Analysis

Now that the dataframe is ready, I will start by getting a general idea of what the processed text of all the Wikipedia pages display. To do this, I will look at the words with the greatest frequencies, or how relevant they are throughout the pages.

```python
#From Notebook 5-2 Topic Modeling
#This is showing the ten most frequent words in the processed text
wiki_str = ' '.join(wiki_data['Processed Text'].tolist())
wiki_tok = wiki_str.split()
wiki_freq = Counter(wiki_tok)
wiki_df = pd.DataFrame(wiki_freq.most_common(20), columns = ["Word", "Frequency"])
wiki_df
```

In terms of the four domains, health seems to be a common word, alluding to culture. Similarly, "development" and "economic" both indicate the economy's impact. Lastly, a word that I found interesting is "happiness." It could relate to health and well-being, in which case it could potentially go under culture, but at the same time, it is a word that can be just as subjective as the quality of life definition itself.

To look at the data more in depth, I will start by making a dataframe just for the processed text, which will later be useful for word embeddings.

```python
wiki_w2v = pd.DataFrame({'Processed': wiki_data["Processed Text"]})
wiki_w2v
```

To identify which domains were most emphasized in these online sources, I created a list of words for each domain, similar to "stopwords," except that they will be counted and not taken out of the text. For consistency, I tried to use Oxford Languages definitions to pick out some key words that relate to each domain, but there are also words that I have included that were not directly in the definition, but still important to the meaning. Lastly, the culture words came from a mix of a definition and the way the Quality Of Life Wikipedia page defines it. This part can vary in terms of what words are used, which is important to keep in mind.

```python
politics_words = "politics", "political", "government", "governments", "governance", "debate", "parties", "power", "diplomacy", "state", "affairs", "political relations", "policy", "law"
economics_words = "economics", "economic", "economy", "production", "consumption", "wealth", "currency", "money", "material", "prosperity"
ecology_words = "ecology", "biology", "environment", "organisms", "physical surroundings", "plants", "animals"
culture_words = "culture", "art", "arts", "intellectual", "achievment", "customs", "social", "belief", "beliefs", "ideas", "creativity", "recreation", "enquiry", "learning", "gender", "generation", "generation", "identity", "engagement", "memory", "projection", "well-being", "wellness", "health", "healthcare"
```

The code below creates a list for each of the domains. It goes through each description under "Processed," and then each word in the description. If the word is in the domain's list of words, then it is appended to the created list. The length of this list will show how many times words related to the domain comes in all the articles, and the Counter / frequencies show which words came up the most.

```python
#politics
pol_words = []
for description in wiki_w2v["Processed"]:
    tokens = description.split()
    for word in tokens:
        if word in politics_words:
            pol_words.append(word)
```

```python
len(pol_words) #how many times words related to politics come up
```

```python
pol_freq = Counter(pol_words)
pol_freq
```

```python
#economics
econ_words = []
for description in wiki_w2v["Processed"]:
    tokens = description.split()
    for word in tokens:
        if word in economics_words:
            econ_words.append(word)
```

```python
len(econ_words)
```

```python
econ_freq = Counter(econ_words)
econ_freq
```

```python
#ecology
ecol_words = []
for description in wiki_w2v["Processed"]:
    tokens = description.split()
    for word in tokens:
        if word in ecology_words:
            ecol_words.append(word)
```

```python
len(ecol_words)
```

```python
#culture
cul_words = []
for description in wiki_w2v["Processed"]:
    tokens = description.split()
    for word in tokens:
        if word in culture_words:
            cul_words.append(word)
```

```python
len(cul_words)
```

According to how often the words relating to each domain come up, the Wikipedia articles seem to emphasize culture and politics the most, suggesting that the quality of life relies on healthcare and wellness, but also what leaders do that impact policy and the lives of citizens. While ecology had fewer words to look for, the word "ecology" itself had only come up sixteen times, less than "culture," "economics," and "politics." Even though the numerical data emphasized more environmental factors, like the Pollution Index, online sources seem to talk more about the other domains.

Through word embedding and word2vec, I will also look at the similarities between the domain words themselves.

```python
split_rows = [row.split() for row in wiki_w2v['Processed']]
```

```python
import multiprocessing 
cpu_count = multiprocessing.cpu_count()
from importlib.metadata import version
```

```python
#From notebook 5-3: Word2vec
model = gensim.models.Word2Vec(split_rows, 
                               vector_size = 100, 
                               min_count = 2, 
                               workers = cpu_count - 1, 
                               window = 3, 
                               sg = 1, 
                               seed = 1) 

words = list(model.wv.index_to_key)
words
```

```python
pol_simwords = model.wv.most_similar("politics", topn=10) 
pol_simwords
```

```python
econ_simwords = model.wv.most_similar("economics", topn=10) 
econ_simwords
```

```python
cul_simwords = model.wv.most_similar("culture", topn=10) 
cul_simwords
```

```python
ecol_simwords = model.wv.most_similar("ecology", topn=10) #can try with "environment" as well
ecol_simwords
```

As can be seen from the similarity values, some of the domains have a connection with each other. For example, "culture" is the most similar word to "politics" and vice versa. This may mean that although the domains have different definitions, they rely upon one another.
